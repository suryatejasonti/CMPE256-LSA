```python
import pandas as pd
import numpy as np
```

```python
df_meta_data = pd.read_csv('metadata-users.csv')
df_meta_data.head()
```

```python
df_test = pd.read_csv('test.csv')
df_test.head()
```

```python
df_train = pd.read_csv('train.csv')
df_train.head()
```

```python
#Import the train_test_split function
from sklearn.model_selection import train_test_split

#Assign X as the original ratings dataframe and y as the user_id column of ratings.
X = df_train.copy()
y = df_train['User']

#Split into training and test datasets, stratified along user_id
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state=42)
```

```python
#Import the mean_squared_error function
from sklearn.metrics import mean_squared_error

#Function that computes the root mean squared error (or RMSE)
def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))
```

```python
#Define the baseline model to always return 3.
def baseline(user_id, track_id):
    return 38
```

```python
#Function to compute the RMSE score obtained on the testing set by a model
def score(cf_model):
    
    #Construct a list of user-movie tuples from the testing dataset
    id_pairs = zip(X_test['User'], X_test['Track'])
    
    #Predict the rating for every user-movie tuple
    y_pred = np.array([cf_model(user, track) for (user, track) in id_pairs])
    
    #Extract the actual ratings given by the users in the test data
    y_true = np.array(X_test['Rating'])
    
    #Return the final RMSE score
    return rmse(y_true, y_pred)
```

```python
score(baseline)
```

```python
#Build the ratings matrix using pivot_table function
r_matrix = X_train.pivot_table(values='Rating', index='User', columns='Track')

r_matrix.head()
```

```python
#User Based Collaborative Filter using Mean Ratings
def cf_user_mean(user_id, track_id):
    
    #Check if movie_id exists in r_matrix
    if track_id in r_matrix:
        #Compute the mean of all the ratings given to the movie
        mean_rating = r_matrix[track_id].mean()
    
    else:
        #Default to a rating of 38 in the absence of any information
        mean_rating = 38
    
    return mean_rating
```

```python
#Compute RMSE for the Mean model
score(cf_user_mean)
```
